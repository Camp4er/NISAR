```python
# Known historical earthquakes in the Shillong Plateau region
# Source: IMD / USGS earthquake catalog

historical_events = [
    {"year": 1897, "magnitude": 8.1, "event": "Great Assam Earthquake", "fault": "Oldham Fault"},
    {"year": 1950, "magnitude": 8.6, "event": "Assam-Tibet Earthquake", "fault": "Mishmi Thrust"},
    {"year": 2009, "magnitude": 5.1, "event": "Bhutan-Assam border", "fault": "MFT"},
    {"year": 2016, "magnitude": 6.7, "event": "Manipur Earthquake", "fault": "Indo-Burma subduction"},
    {"year": 2021, "magnitude": 6.4, "event": "Sonitpur, Assam", "fault": "Kopili Fault"},
    {"year": 2023, "magnitude": 4.8, "event": "Meghalaya tremor", "fault": "Oldham Fault"},
]

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

years = [e["year"] for e in historical_events]
mags  = [e["magnitude"] for e in historical_events]
names = [e["event"] for e in historical_events]

fig, ax = plt.subplots(figsize=(13, 5))
colors = ['#e74c3c' if m >= 7 else '#e67e22' if m >= 6 else '#3498db' for m in mags]
bars = ax.bar(range(len(years)), mags, color=colors, edgecolor='white', linewidth=0.8)

ax.set_xticks(range(len(years)))
ax.set_xticklabels([f"{y}\n{n}" for y, n in zip(years, names)], fontsize=8, rotation=15, ha='right')
ax.set_ylabel('Magnitude', fontsize=11)
ax.set_title('Major Seismic Events — Shillong Plateau & NE India\n(Oldham Fault & surrounding systems)', fontsize=13, fontweight='bold')
ax.set_ylim(0, 10)
ax.axhline(y=7, color='red', linestyle='--', alpha=0.4, label='M7.0 threshold')
ax.axhline(y=6, color='orange', linestyle='--', alpha=0.4, label='M6.0 threshold')

legend = [
    mpatches.Patch(color='#e74c3c', label='Major (M≥7)'),
    mpatches.Patch(color='#e67e22', label='Significant (M≥6)'),
    mpatches.Patch(color='#3498db', label='Moderate (M<6)'),
]
ax.legend(handles=legend, loc='upper right')
plt.tight_layout()
plt.savefig(r"C:\Users\saxen\OneDrive\Desktop\NISAR_Project\outputs\historical_earthquakes.png", dpi=150)
plt.show()
print("✅ Historical comparison chart saved!")
```
